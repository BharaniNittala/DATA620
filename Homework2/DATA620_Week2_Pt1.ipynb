```python
import networkx as nx
import pandas as pd
```

Create an empty graph

```python
g = nx.Graph()
```

Import csv file

```python
df = pd.read_csv('com-amazon.top5000.cmty.csv',header=None)
```

```python
df.head()
```

Iterate over each row of the Pandas dataframe to add each row as an edge to graph 'g'

```python
for index, row in df.iterrows():
    g.add_edge(row[0], row[0])
```

```python
g.edges()
```

```python
g.nodes()
```

Create graph

```python
nx.draw(g)
```
